### Imports

```python
import pandas as pd 
# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (22,5)
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

### Load Data

```python
# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 
vre_profiles_2006_data = pd.read_csv("inputs/vre_profiles2006.csv", names=["Type", "Hour", "Production Profile"])
vre_profiles_2006_data = vre_profiles_2006_data.pivot(index='Hour', columns='Type')
# Preview the first 5 lines of the loaded data 
vre_profiles_2006_data.describe()
```

```python

# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 
demand_2006_data = pd.read_csv("inputs/demand2006.csv", names=["Hour", "Demand"])
# Preview the first 5 lines of the loaded data 
demand_2006_data.head()
```

```python
def plotProfile(profileName):
    f, ax = plt.subplots()
    ax.plot(vre_profiles_2006_data["Production Profile"][profileName])
    ax.set(xlabel='Hour', ylabel='Production profile', title=profileName + 'Offshore electricity â€“ Yearly production profile')
    ax.grid()
    mean=vre_profiles_2006_data["Production Profile"][profileName].mean()
    plt.axhline(mean, color='r', linestyle='--')
    plt.legend(['Production', 'Mean'])
    plt.ylim(0, 1)
    plt.show()
  
```

```python
def plotDemand():
    f, ax = plt.subplots()
    ax.plot(demand_2006_data['Hour'], demand_2006_data['Demand'])
    ax.set(xlabel='Hour', ylabel='Demand',
       title='Electricity Demand (2006)')
    ax.grid()
    plt.show()
```

```python
plotProfile('onshore')
plotProfile('offshore')
plotProfile('pv')
plotDemand()
```
